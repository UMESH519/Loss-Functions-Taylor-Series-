### Q1 : Loss Functions (30 points)

Implement the following loss functions using numpy (each function should take two numpy arrays as arguments and return the value of the loss):

1. Huber Loss (10 points)
2. Hinge Loss (10 points)
3. Cross Entropy Loss (10 points)

```python
import numpy as np
def huber_loss(predictions,label,delta):
    c=np.where(abs(predictions-label)<delta,1/2*((predictions-label)**2),delta*(abs(predictions-label))-1/2*delta**2)
    return np.mean(c)

predictions =np.array([3,4,5,6,7])
labels =np.array([4,9,9,5,10])
huber_loss(predictions,labels,3)

predictions =np.array([3,4,5,6,7])
labels =np.array([4,9,9,5,10])
huber_loss(predictions,labels,5)    
    
```

```python

def hinge_loss(predictions,label):
    c=np.maximum(0,(1-predictions*label))
    return np.mean(c)
    

predictions = np.array([-0.3,-0.5,6.6,0.2,0.9])
labels = np.array([1,-1,-1,1,1])
hinge_loss(predictions,labels)
        
        
    
```

```python
def cross_entropy_loss(predictions,labels):
    N=predictions.shape[0]
    error=-1/N*(np.sum(np.where(labels==1,np.log(predictions+1e-20),np.log(1-predictions+1e-20))))
    return error

predictions=np.array([0.2,0.5,0,0.7,0.05])
labels=np.array([1,0,0,1,1])
cross_entropy_loss(predictions,labels)
```

### Q2 : Visualizing Taylor Series (10 points)

Taylor series for a function $f(x)$ about $x=a$ is :
$$\begin{align*}f\left( x \right) & = \sum\limits_{n = 0}^\infty {\frac{{{f^{\left( n \right)}}\left( a \right)}}{{n!}}{{\left( {x - a} \right)}^n}} \\ & = f\left( a \right) + f'\left( a \right)\left( {x - a} \right) + \frac{{f''\left( a \right)}}{{2!}}{\left( {x - a} \right)^2} + \frac{{f'''\left( a \right)}}{{3!}}{\left( {x - a} \right)^3} + \cdots \end{align*}$$

Visualize the 1st, 2nd and 4th order Taylor series approximation for cos(x) at $x=0$


```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(10,8))
x=np.arange(-2*np.pi,+2*np.pi,0.1)
y=np.cos(x)

plt.plot(x,y)
#for degree 1
y1=np.cos(0) - np.sin(0)*(x) 
plt.plot(x,y1)

#for degree 2
y2=(y1 - (np.cos(0)*((x-0)**2))/2)
plt.plot(x,y2)

#for degree 4
y4=y2 + 0 + (np.cos(0)*((x-0)**4))/np.math.factorial(4)

plt.plot(x,y4)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.legend(["Cos x Fucntion","Degree 1","Degree 2","Degree 4"])
plt.title("Approximation using Taylor Series")
plt.ylim(-4,4)
```

### Q3 : Visualising Loss surface in weight space (20 points)

Consider a loss function defined in terms of weights w1 and w2 as : $$ J = \frac{1}{2}[(w_1 - w_2)^2 + (25 - w_1)^2]$$

1. Plot the loss surface in the weight space $w_1$ and $w_2$  (10 points)
2. At any three points, plot the direction of the gradient at that point (10 points)

```python
import numpy as np
import matplotlib.pyplot as plt
w1=np.arange(-100,100,5)
w2=np.arange(-100,100,5)
[X, Y] = np.meshgrid(w1, w2)
J=1/2*np.array((X-Y)**2+(25-X)**2)

fig,ax=plt.subplots(1,1)
ax.contour(X, Y, J,levels=50)
ax.quiver(1,1,-24,0)
ax.quiver(-10,20,-85,30)
ax.quiver(50,50,25,0)
ax.set_xlabel('w1')
ax.set_ylabel('w2')
ax.set_title('Loss function J')
fig.colorbar(ax.contour(X, Y, J,levels=50))


```
